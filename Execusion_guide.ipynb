The following code generates figures 1-4 of the paper "Recovery Guarantees for Distributed-OMP", accepted for publiaction at AISTATS 2024. There are four stages to each simulation:
1. Setup simulation parameters
2. Execute parallel simulations on a computing cluster
3. Aggregate results
4. Plot figures

For example, let's generate the first figure. To this end, we begin with running the appropriate setup file (this might take a while):


```python
#Step 1 - Setup simulation parameters

#Please make sure that you are in the correct directory before running this cell
import os
#os.chdir(<code directory path>)

%run setup_fig1a.py 
```

Once setup is completed, we should have a folder named "setting1a" contating a pars.dat file and subfolders containing X matrices and mvXt debiasing terms, as well as additional auxillary files. 
    The next step is to submit jobs to a computing cluster that executes muliple copies of run_exp.py in parallel. This python file receives two command line arguments - the first is setting_id which indicates which setting is being simulated (e.g., 1a), and the second is task_id, an id for the independent noise realization (e.g., 21). To run this file, the cluster should have installations of python (we used v3.8) and R (we used v4.0.3). 

```python
#Example of running one realization

#Please make sure that you are in the correct directory before running this cell
import os
#os.chdir(<code directory path>)

%run run_exp.py 1a 21
```

Once the simulations are completed, we should have a subfolder named "results" with the simulation results. Now we can use the following code to aggragate them and to plot the figures.

```python
#Step 3 - Aggregate results (setting 1a) 
import os
import pickle
from helpers import pic_load, pic_save, aggregate_results
import numpy as np



#os.chdir(<code directory path>)
data_folder = './setting1a/'

#load parameters
pars_saved = pic_load(data_folder + 'pars.dat')
for key, val in pars_saved.items():
    exec(key + '=val')

NumOft_min = len(t_min_Vec)
NumOfd = len(d_Vec)
NumOfM = len(M_Vec)   
NumOfK = len(K_Vec)
NumOfAlgs = len(alg_names)

Succ_mat, Duration_mat = aggregate_results(data_folder, NumOfd, NumOfM, NumOfK, NumOft_min, NumOfAlgs, NumOfSig)
Succ_mean = np.mean(Succ_mat,axis=-1)
```

```python
#Step 4 - Plot (figure 1a)
import matplotlib.pyplot as plt
import matplotlib as mpl
from cycler import cycler
from helpers import alg_names_for_display

plt.rcParams.update({'font.size': 50})
plt.figure(figsize=(17,13))

mpl.rcParams['text.usetex'] = True

alg_order = [4, 5, 6, 7, 0, 2, 3, 1]
mpl.rcParams['axes.prop_cycle'] = cycler(
                color=['red', 'orange', 'gray', 'green', 'blue', 'brown','pink' , 'purple'])

marker_list = ['s', '+', 'x', 'o', '1', 'd', '*', '^']

d_Ind=0
d=d_Vec[d_Ind]
M_Ind=0
M=M_Vec[M_Ind]
K_Ind=0
K=K_Vec[K_Ind]
X_dir_name = data_folder +'/X_d_Ind%d/' %(d_Ind)
mumax = np.load(X_dir_name + 'mumax'+str(M)+'.npy')
mu_stacked = np.load(X_dir_name + 'mu_stacked'+str(M)+'.npy')

succ_plot = []
for alg_num, alg_Ind in enumerate(alg_order):
    line = plt.plot(t_min_Vec, Succ_mean[d_Ind,M_Ind,K_Ind,:, alg_Ind],
                     label=alg_names_for_display(alg_names[alg_Ind]), 
                     linewidth=8, 
                     marker=marker_list[alg_num],
                     markersize=25)
    succ_plot.append(line)
plt.rcParams.update({'lines.markeredgewidth': 10})

IT_LB = 1/(M*n)
dom = 1-(2*K-1)*mu_stacked
if dom>=0:
    IT_LB_tmin = np.sqrt(2*IT_LB*np.log(d-K)/dom)
    line = plt.axvline(x=IT_LB_tmin, linestyle='--', color='black', linewidth=8)
    succ_plot.append(line)


plt.ylim([0, 1])
plt.xlim([0.02, 0.1])
plt.xlabel(r"$\theta_{\min}$")
plt.ylabel("Pr[Exact Support Recovery]")
plt.grid(True)

plt.savefig(data_folder + 'fig1a.eps', format='eps')
plt.show()
plt.clf()
```

```python
#Step 3 - Aggregate results (setting 4) 
import os
import pickle
from helpers import pic_load, pic_save, aggregate_results
import numpy as np



#os.chdir(<code directory path>)
data_folder = './setting4/'

#load parameters
pars_saved = pic_load(data_folder + 'pars.dat')
for key, val in pars_saved.items():
    exec(key + '=val')

NumOft_min = len(t_min_Vec)
NumOfd = len(d_Vec)
NumOfM = len(M_Vec)   
NumOfK = len(K_Vec)
NumOfAlgs = len(alg_names)


Succ_mat, Duration_mat = aggregate_results(data_folder, NumOfd, NumOfM, NumOfK, NumOft_min, NumOfAlgs, NumOfSig)
Succ_mean = np.mean(Succ_mat,axis=-1)


t_min_Ind = 0
alg_Ind = alg_names.index('DJ-OMP') 

pr_value = 0.5
estimated_M = np.zeros([NumOfd,NumOfK])

for d_Ind in range(len(d_Vec)):
    for K_Ind in range(len(K_Vec)):
        Succ_per_M = Succ_mean[d_Ind,:,K_Ind,t_min_Ind,alg_Ind]
        end = np.argmax(Succ_per_M>=pr_value)
        start=end-1
        if end>0:
            estimated_M[d_Ind,K_Ind] = (pr_value-Succ_per_M[start])*(M_Vec[end]-M_Vec[start])/(Succ_per_M[end]-Succ_per_M[start])+M_Vec[start]


```

```python
#Step 4 - Plot (figure 4)
import matplotlib.pyplot as plt
import matplotlib as mpl
from cycler import cycler
from helpers import alg_names_for_display
import matplotlib.ticker as ticker
from matplotlib.ticker import ScalarFormatter

plt.rcParams.update({'font.size': 50})
plt.figure(figsize=(17,13))

mpl.rcParams['text.usetex'] = True

mpl.rcParams['axes.prop_cycle'] = cycler(
                color=['red', 'orange', 'green'])

marker_list = ['s', 'o',  'd']

M_vs_d_plot = []
for K_Ind, K in enumerate(K_Vec):
    scat = plt.scatter(d_Vec, estimated_M[:,K_Ind], 
                label=("K=%d"%K),
               marker=marker_list[K_Ind])
    M_vs_d_plot.append(scat)
    c, stats = np.polynomial.polynomial.polyfit(np.log(d_Vec), np.log(estimated_M[:,K_Ind]), 1, full=True)
    yy = np.exp(np.multiply(c[1],np.log(d_Vec))+c[0])
    line = plt.plot(d_Vec, yy, '--',
             label=(r"Linear fit, $\beta$=%.3f, SSR=%.4f"%(c[1],stats[0][0]))
            )
    M_vs_d_plot.append(line)
plt.rcParams['lines.markersize'] = 25


plt.yscale('log')
plt.xscale('log')
plt.xlabel(r"$d$")
plt.ylabel(r"Estimated $M$")

plt.rc('legend',fontsize='36')
legend_order = [0,2,4,1,3,5]
handles, labels = plt.gca().get_legend_handles_labels()
plt.legend([handles[idx] for idx in legend_order],[labels[idx] for idx in legend_order],ncol=2,loc='upper center', bbox_to_anchor=(0.5, -0.25))
plt.tight_layout()

plt.grid(True)

ax = plt.gca()
ax.xaxis.set_major_formatter(ScalarFormatter())
ax.xaxis.set_minor_formatter(ScalarFormatter())
plt.gca().xaxis.set_major_locator(plt.NullLocator())
plt.gca().xaxis.set_minor_locator(plt.NullLocator())


ax.yaxis.set_major_formatter(ScalarFormatter())
ax.yaxis.set_minor_formatter(ScalarFormatter())
plt.gca().yaxis.set_major_locator(plt.NullLocator())
plt.gca().yaxis.set_minor_locator(plt.NullLocator())
plt.yticks(M_Vec)
plt.xticks(d_Vec)

plt.savefig('fig4.eps', format='eps')
plt.show()
plt.clf()
```
